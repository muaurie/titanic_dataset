```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('titanic.csv')
#cereate list of Age values no including N/A values
ls_age = df['Age'].dropna()
#plot the data in this list into a histogram
sns.displot(ls_age, hist=True, kde=False, bins=16)
# What does kde stand for? https://seaborn.pydata.org/generated/seaborn.distplot.html
#The histogram is the value at an interval of the x-axis, and the PDF y-axis is the probability of being that value.
sns.distplot(df['Age'].dropna(), hist=True, kde=True, bins=16)
#create the CDF of the PDF we did in the Age feature
sns.distplot(df['Age'].dropna(), hist_kws=dict(cumulative=True), kde_kws=dict(cumulative=True))
#Over 85% of the passengers of the Titanic were over 18
```
